### Funciones en python
Una función en python es un objeto que se puede utilizar posteriormente
* Dentro de los parentesis estaran los argumentos de la función, es decir, que es lo que recive la función como input
* la función puede tener o no la clausula return, que indica que es lo que la función nos retorna

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Para declarar una función es necesario la clausula def
def f(x): # la función recive 2 argumentos
    return x**5+x**4+2*x**3+x**2+2*x+5
    
x = np.arange(-50000,50000)/10
fx = f(x)

plt.plot(x,fx)
```

### if en python
If es una clausula (o un bloque de programación) que se ejecutará si se cumple una condición

```python
x = 0
if x == 0:
    print("Hola!")
```

### else en python
else es una sentencia de python que se puede (o no) poner detras del if, se ejecutará esa parte de código si la condición del if no se cumple

```python
x = -1
if x == 0:
    print("x es 0, estoy dentro del if")
else:
    print("x es diferente de 0, estoy en el else!")
```
