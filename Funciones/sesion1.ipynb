## Apuntes tipo de variables booleanas
En python existe un tipo de objeto llamado booleano, y nos dice si algo es cierto o es falso

Operaciones de comparación:
   * <, >, <=, >=: Comparan si un elemento es mayor, menor, mayor igual o menor igual que otro elemento. Ej 3 > 2 => False
   * ==: Compara si un elemento es igual a otro (Cuidado! es muy fácil confundirse con =, el cual utilizamos para asignar un valor a una variable)
   * !=: compara si dos elementos no son iguales
   
#### Ejemplos

```python
print("2 > 4 es", 2 > 4) # esto es Falso
print("3 <= 4 es", 3 <= 4) # esto es Verdadero
print("2+2 == 4 es", 2+2 == 4) # Verdadero, e prioriza la operación numerica a la de comparación
x=3
y=8
print("x es diferente de y es", x != y)
```

Operaciones con booleanos:
   * and: compara que dos elementos booleanos son verdaderos. Ej: True and False => False
   * or: compara que de dos elementos booleanos haya uno verdadero. Ej: True or True => True
   * not: cambia el valor de un booleano. Ej: not True => False
 
#### Ejemplos

```python
print("Verdadero Y Falso es ", True and False)
print("Verdadero Y Verdadero es ", True and True)
print("Verdadero O False es ", True or True)
print("Falso O Falso es ", False or False)
print("No Falso es", not False)
```

```python
x = 3
y = 5
print("x == 3 Y y == 5 es", x == 3 and y == 5)
print("x < y o x == 999 es", x < y or x == 999)
```

```python
# voy a escribir una expresion booleana y guardar el resultado en una variable
booleano = 3 > 5 or 6 > 1
print("Mi variable booleana es", booleano)
```
