### Funciones en python
Una función en python es un objeto que se puede utilizar posteriormente
* Dentro de los parentesis estaran los argumentos de la función, es decir, que es lo que recive la función como input
* la función puede tener o no la clausula return, que indica que es lo que la función nos retorna

```python
# Para declarar una función es necesario la clausula def
def mi_funcion(): # mi_funcion es el nombre de la funcion
    print("hola!") # es necesario que el contenido de la función este tabulado
    
# para utilizar la funcion
mi_funcion()
```

```python
# Para declarar una función es necesario la clausula def
def mi_funcion(x, y): # la función recive 2 argumentos
    return x*y
    
# para utilizar la funcion
print(mi_funcion(3,5))

# si hay un return, la función retorna un objeto (un valor)
x = mi_funcion(3,5)
print("El resultado de mi funcio es", x) 

# la función se puede utilizar directamente como un resultado
print("El resultado de 3 por 5 es 20, esta frase es", mi_funcion(3,5)==20)
```

```python
def mi_funcion(m): # la función recive 1 argumento, pero m no existe fuera de la función
    return m**2

print(mi_funcion(4))

# esto nos retorna un error porque m no existe fuera de la función
print(m)
```

```python
# Dentro de las funciones pueden haber los otros elementos de python
def tabla_multiplicar(m):
    for i in range(11):
        print(i, "por", m, "es", i*m)

tabla_multiplicar(4)
```

### Gráficas en Python
Matplotlib es una libreria para plots en python

```python
# importamos la libreria
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# la función plot es la función básica para graficar
# el primer argumento son los puntos del eje horizontal (x)
# el segundo argumento son los puntos del eje vertical (y)
plt.plot([1,2,3],[6,2,4]) 
```

```python
v = np.arange(-1000,1000)
cuadrado = v**2

plt.title('Titulo') # titulo
plt.ylim((0,10000)) # con ylim podemos definir el rango del eje vertical
plt.xlim((-50,100))# con xlim podemos definir el rango del eje horizontal
plt.plot(v,cuadrado)
```

```python
plt.title('Titulo') # titulo
plt.ylim((0,100)) # con ylim podemos definir el rango del eje vertical
plt.xlim((-5,10))# con xlim podemos definir el rango del eje horizontal

plt.plot(v, cuadrado, color = "r") # podemos poner diferentes lineas en un mismo grafico
plt.plot(v, v, color = "g") # con diferente color

# podemos poner el nombre de los ejes
plt.xlabel("X")
plt.ylabel("f(x)")
plt.show() # con esto mostramos el grafico sin el Out
```

### Ejercicio 3: Gráficas
La siguiente función se llama función de Weierstrass. No hay que modificar nada de la función.

```python
def weierstrass(x, a, b):
    suma = np.zeros(len(x))
    for i in range(10):
        suma = suma + a**i*np.cos(x*np.pi*b**i)
    return suma
```

a) Gráfica la función de Weierstrass, primero crea un array de 1000 valores entre el -0.5 y 0.5. Después pasa a la función el array como x, $a=0.7$ y $b=3$.

```python
# Solución
x=np.arange(-500,500)/1000
fx=weierstrass(x, 0.7, 3)
plt.plot(x, fx, color = "r")
```
