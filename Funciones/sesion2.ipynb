### Bucles fors
Un bucle en python es una instrucción que se repite varias veces

Los fors son un bucle que en cada iteración toma el valor de una lista

La instrucción que se hace dentro del for debe estar tabulada (hay que presionar la tecla tab del teclado para identar)

```python
# range(n) es una función que retorna numeros del 0 al n (n entero)
# i es el nombre de a variable que toma el valor de 0 a 10
for i in range(10):
    print("En esta iteración, el valor de i es", i)
```

```python
# Si utilizamos otro argumento en el range, nos dira desde donde empieza el valor de i
for i in range(5,10): # empezamos en 5 y llegamos hasta 10(sin incluirlo)
    print("En esta iteración, el valor de i es", i)
```

```python
# Si utilizamos otro argumento en el range(tercer argumento), indicara el intervalo entre numeros
for i in range(0, 10, 2): # contamos de 2 en 2, desde 0 hasta 10(sin incluirlo)
    print("En esta iteración, el valor de i es", i)
```

##### Ejemplos
Podemos sumar numeros del 0 al 9

```python
suma = 0 # declaro una variable que sera el resultado de toda la suma
for i in range(10): # hago un bucle for que va desde i = 0 hasta i = 9
    suma = suma + i # voy acumulando la suma
print(suma)
```

### Instrucciones básicas Numpy
Numpy es una libreria para operar con arrays (listas de numeros)

```python
# cargo la libreria y le pongo un nombre corto (np)
import numpy as np
```

```python
# declaro un array(vector) y lo guardo en una variable v
v = np.array([1,2,3,4,5])
print(v)

# este array tiene una longitud, podemos saber cual es mediante la funcion len(array)
print("La longitud de mi vector es", len(v))

# declaro un vector de 0 con 10 posiciones
v = np.zeros(10)
print(v)

# declaro un vector de 1 con 10 posiciones
v = np.ones(10)
print(v)

# declaro un vector del 0 al 10 (sin incluir el 10)
v = np.arange(10) # np.arange se comporta de forma parecida al range de los fors
print(v)

# np.arange se comporta de forma parecida al range de los fors
v = np.arange(10,20,2) # empiezo en 10 hasta 20 y salto de 2 en 2
print(v)

# puedo acceder a la información guardada en las posiciones del array con []. 
# Dentro de [] va el numero de posición: de 0 a la longitud -1
print(v[2]) 

# las posiciones empiezan en 0
print(v[0])

# la ultima posición es len(n)-1
print(v[len(v)-1])

# puedo cambiar el valor de la posición 2 (tercera posición) del array
v[2] = 99
print(v)
```

```python
# declaro un array
v = np.arange(10)

# se puede operar facilmente todo el vector
print(v**2)
print(2*v)
print(v/10)

# se pueden sumar y restar vectores de la misma longitud
print(v+2*v+v**2)
```

```python
# rellenar XXX y YYY con ejemplos propios
v=np.arange(10)
suma=np.zeros(len(v))
for i in range(1, len(v)):
    suma[i] = suma[i-1]+v[i]
print(suma)
```
