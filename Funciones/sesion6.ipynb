### Funciones y gráficas en Python
Una función en python es un objeto que se puede utilizar posteriormente
* Dentro de los parentesis estaran los argumentos de la función, es decir, que es lo que recive la función como input
* la función puede tener o no la clausula return, que indica que es lo que la función nos retorna

```python
# Cargamos librerias
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Para declarar una función es necesario la clausula def
def f(x): # la función recive 2 argumentos
    return x**5+x**4+2*x**3+x**2+2*x+5
    
x = np.arange(-50000,50000)/10
fx = f(x)

plt.title('Titulo') # titulo
plt.ylim((-100,100)) # con ylim podemos definir el rango del eje vertical
plt.xlim((-5,5))# con xlim podemos definir el rango del eje horizontal

plt.plot(x, fx, color = "r") # podemos poner diferentes lineas en un mismo grafico
plt.plot(x, x, color = "g") # con diferente color

# podemos poner el nombre de los ejes
plt.xlabel("X")
plt.ylabel("f(x)")
plt.show() # con esto mostramos el grafico sin el Out
```

### Ejercicio 2: Funciones Racionales, dominio e imagen 2
a) Ejercicio entregable. Programa la siguiente función. ¿Cuál es el dominio? ¿Y la imagen? ¿En que puntos corta con los ejes?
$$ f(x)=   \left\{
\begin{array}{ll}
      \frac{-x^4}{x^2+2x+3} & x < 0 \\
      \frac{1}{x-5} & 0 \leq x  \\
\end{array} 
\right.  $$

```python
# Definición de la función
def f(x):
    if x < 0:
        return -x**4/(x**2+2*x+3)
    else:
        return 1/(x-5)
```

```python
x= np.arange(-1000,1000)/100
fx=np.zeros(len(x))
for i in range(len(x)):
    fx[i] = f(x[i])
    
plt.title('Titulo') # titulo
plt.ylim((-100,100)) # con ylim podemos definir el rango del eje vertical
plt.xlim((-10,10))# con xlim podemos definir el rango del eje horizontal

plt.plot(x, fx, color = "r") # podemos poner diferentes lineas en un mismo grafico
plt.plot(x, np.zeros(len(x)), color = "g")
plt.plot(np.zeros(len(x)), x, color = "g")
# podemos poner el nombre de los ejes
plt.xlabel("X")
plt.ylabel("f(x)")
plt.show() # con esto mostramos el grafico sin el Out
```

### Ejercicio 3: Problemas de funciones racionales

Ejercicios entregables a y b

Sea $f(x)$ una función descrita de la siguiente forma:
$$
f(x) = \frac{h(x)}{g(x)}
$$

Donde $h(x)$ y $g(x)$ son funciones que desconocemos por el momento. En python esto se programa así


```python
def f(x, g, h): # se pueden utilizar funciones como argumento y utilizarlas dentro:
    return h(x)/g(x)

# Siempre en el caso que h y g no tengan más argumentos
```

a) Ejercicio entregable. Programa las funciones siguientes:

$$
h(x) = (x-5)^2\\
g(x) = 2x
$$

Y grafica la función $f(x)$. ¿Cuál es el dominio? ¿Y la imagen? ¿En que puntos corta con los ejes?

```python
def h(x):
    return (x-5)**2
def g(x):
    return 2*x
```

```python
x= np.arange(-1000,1000)/100
fx=np.zeros(len(x))
for i in range(len(x)):
    fx[i] = f(x[i], g, h)
    
plt.title('Titulo') # titulo
plt.ylim((-100,100)) # con ylim podemos definir el rango del eje vertical
plt.xlim((-10,10))# con xlim podemos definir el rango del eje horizontal

plt.plot(x, fx, color = "r") # podemos poner diferentes lineas en un mismo grafico
plt.plot(x, np.zeros(len(x)), color = "g")
plt.plot(np.zeros(len(x)), x, color = "g")
# podemos poner el nombre de los ejes
plt.xlabel("X")
plt.ylabel("f(x)")
plt.show() # con esto mostramos el grafico sin el Out
```
