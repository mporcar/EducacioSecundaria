### Funciones en python
Una función en python es un objeto que se puede utilizar posteriormente
* Dentro de los parentesis estaran los argumentos de la función, es decir, que es lo que recive la función como input
* la función puede tener o no la clausula return, que indica que es lo que la función nos retorna

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Para declarar una función es necesario la clausula def
def f(x): # la función recive 2 argumentos
    return x**5+x**4+2*x**3+x**2+2*x+5
    
x = np.arange(-50000,50000)/10
fx = f(x)

plt.title('Titulo') # titulo
plt.ylim((-100,100)) # con ylim podemos definir el rango del eje vertical
plt.xlim((-5,5))# con xlim podemos definir el rango del eje horizontal

plt.plot(x, fx, color = "r") # podemos poner diferentes lineas en un mismo grafico
plt.plot(x, x, color = "g") # con diferente color

# podemos poner el nombre de los ejes
plt.xlabel("X")
plt.ylabel("f(x)")
plt.show() # con esto mostramos el grafico sin el Out
```

### else en python
else es una sentencia de python que se puede (o no) poner detras del if, se ejecutará esa parte de código si la condición del if no se cumple

```python
x = -1
if x == 0:
    print("x es 0, estoy dentro del if")
else:
    print("x es diferente de 0, estoy en el else!")
```

### Ejercicio 3: Problema de funciones definidas a trozos
Ejercicio enntregable a, b y c

a) El coste del internet en una compañia de telefonía es de 30 euros de base hasta que se supera cierto limite de consumo $L$ de GB. En el momento que se supera el consumo $L$ cada GB extra cuesta 0.5 euros. Programad la función de coste por consumo en función de L.

```python
def f(x, L):
    if x < L:
        return 30
    else:
        return 30+0.5*(x-L)
```

```python
x = np.arange(70)
fx = np.zeros(len(x))
for i in range(len(x)):
    fx[i] = f(x[i], 20)
    
plt.title('Coste por consumo') # titulo
plt.ylim((0,100)) # con ylim podemos definir el rango del eje vertical
plt.xlim((0,70))# con xlim podemos definir el rango del eje horizontal

plt.plot(x, fx, color = "r") # podemos poner diferentes lineas en un mismo grafico

# podemos poner el nombre de los ejes
plt.xlabel("GB")
plt.ylabel("Euros")
plt.show() # con esto mostramos el grafico sin el Out
```

c) La compañia ha detectado que hay un grupo de consumidores que consumen muchísimo internet y quieren penalizar este consumo. Para ello, cuando el consumo supere $2L$, el coste del margen de GBs que superen el limite $2L$ será el margen del GBs al cuadrado. Nota: En el tercer margen de la función se debe agregar el coste de los GB del tramo [$L$, $2L$] que simplemente es 0.5$L$.

```python
def f(x, L):
    if x < L:
        return 30
    if x>=L and x <2*L:
        return 30+0.5*(x-L)
    if x >= 2*L:
        return 30+0.5*L+(x-2*L)**2
    
    
x = np.arange(70)
fx = np.zeros(len(x))
for i in range(len(x)):
    fx[i] = f(x[i], 20)
    
plt.title('Coste por consumo') # titulo
plt.ylim((0,100)) # con ylim podemos definir el rango del eje vertical
plt.xlim((0,70))# con xlim podemos definir el rango del eje horizontal

plt.plot(x, fx, color = "r") # podemos poner diferentes lineas en un mismo grafico

# podemos poner el nombre de los ejes
plt.xlabel("GB")
plt.ylabel("Euros")
plt.show() # con esto mostramos el grafico sin el Out
```
